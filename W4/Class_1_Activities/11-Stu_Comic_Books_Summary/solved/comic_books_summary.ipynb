```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# File to Load
comic_path = 'Resources/books_clean.csv'

# Read the modified Comic Books csv and store into Pandas DataFrame
comic_df = pd.read_csv(comic_path)
comic_df
```

```python
# Calculate the number of unique authors in the DataFrame
num_authors = len(comic_df['Author'].unique())

# Calculate the number of unique publication countries in the DataFrame
num_countries = len(comic_df['Country of Publication'].unique())

# Print the number of unique authors and countries. 
print(num_authors)
print(num_countries)
```

```python
# Calculate the earliest/latest year a book was published
early_yr = comic_df['Publication Year'].min()
late_yr = comic_df['Publication Year'].max()

# Print the earliest and latest year of publication.
print(early_yr)
print(late_yr)
```

```python
# Place all of the data found into a summary DataFrame
comic_summ_df = pd.DataFrame([
    {'Total Unique Authors' : num_authors, 'Total Unique Publication Countries' : num_countries, 'Earliest Year' : early_yr, 'Latest Year' : late_yr}
]
)
comic_summ_df
```
