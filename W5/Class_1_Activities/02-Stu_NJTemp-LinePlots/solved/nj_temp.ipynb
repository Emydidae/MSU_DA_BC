```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x = np.arange(1, 13, 1)
x
```

```python
# Average weather temp
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Plot the line
plt.plot(x, points)
plt.show()
```

```python
# Convert to Celsius C = (F-32) * 0.56
c_points = [(point - 32) * 0.56 for point in points]
c_points
```

```python
# Plot using Celsius
plt.plot(x, c_points)
plt.show()
```

```python
# Plot both on the same chart
plt.plot(x, points, c_points)
plt.show()
```
