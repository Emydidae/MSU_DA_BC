```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
state_df = pd.read_csv('../Resources/avg_rain_state.csv')
state_df.head()
```

```python
# Set x axis and tick locations
x = np.arange(len(state_df['State']))
plt.bar(x, state_df['Inches'], color='r', alpha=0.5, align='center')
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.xticks(x, state_df['State'], rotation=90)
plt.rcParams['figure.figsize'] = [20,4]
# Set x and y limits
plt.xlim(-.75, len(x)-.25)
plt.ylim(0, 70)
# Set a Title and labels
plt.title('Average Rain per State')
plt.xlabel('State')
plt.ylabel('Average Amount of Rain (In)')
# Save our graph and show the graph
plt.savefig('avg_state_rain.png')
plt.show()
```
