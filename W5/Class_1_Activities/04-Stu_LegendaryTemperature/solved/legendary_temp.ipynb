```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x = np.arange(1, 13, 1)
x
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = [((f - 32) * .56) for f in points_F]
points_C
```

```python
# Create a handle for each plot
temp_F, = plt.plot(x, points_F, marker = '+', color = 'blue', label = 'Fahrenheit')
temp_C, = plt.plot(x, points_C, marker = 's', color = 'red', label = 'Celsius')

# Set our legend to where the chart thinks is best
plt.legend(handles=[temp_F,temp_C], loc='best')

# Create labels for the X and Y axis
plt.xlabel('Month')
plt.ylabel('Temperature')

# Save and display the chart
plt.savefig('temp_chart.png')
plt.show()
```
