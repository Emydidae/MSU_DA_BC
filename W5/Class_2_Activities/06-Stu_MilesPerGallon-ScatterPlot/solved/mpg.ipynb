```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('../Resources/mpg.csv')
car_data.head()
```

```python
# Remove the rows with missing values in horsepower
car_data_clean = car_data[car_data['horsepower'] != '?']
car_data_clean.head()
```

```python
# Set the 'car name' as our index
car_data_clean = car_data_clean.set_index('car name')
# Remove the 'origin' column
del car_data_clean['origin']
car_data_clean.head()
```

```python
# Convert the "horsepower" column to numeric so the data can be used
car_data_clean = car_data_clean.astype({'horsepower':'float'})
car_data_clean.dtypes
```

```python
# Create a scatter plot which compares MPG to horsepower
fig = car_data_clean.plot(kind = 'scatter', x = 'horsepower', y = 'mpg', title = 'MPG Vs. Horsepower')
```
