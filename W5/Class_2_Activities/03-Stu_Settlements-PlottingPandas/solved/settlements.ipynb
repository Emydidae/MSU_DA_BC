```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
df = pd.read_csv('Resources/union_settlements_1995.csv')
df.head()
```

```python
# List columns
df.columns
```

```python
# Get total settlements by union
union_data = df['UNION'].value_counts()
# union_df = df.groupby(['UNION'])
# union_count = union_df.count()
# union_count = union_count.sort_values('SCHEDULE NUMBER',
#                                       ascending = False)
# union_count
```

```python
# Configure plot, figsize, title, and axis labels
union_count_plot = union_data.plot(kind='bar', figsize=(8,6), facecolor='red', title='Major Collective Bargaining Settlements (1995)', xlabel='Union', ylabel='Settlements')

# Configure x-tick rotation
labels = union_data.index
union_count_plot = union_count_plot.set_xticklabels(labels, rotation = 45, rotation_mode='anchor', ha='right', wrap=True)
# Show plot
union_count_plot
```
